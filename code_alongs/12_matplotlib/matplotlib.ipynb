# matplotlib
- focus on the distinctionn between two ways of writing matplotlib codes
- focus on the components of a matplotlib figure 

```python
import matplotlib.pyplot as plt
import numpy as np

# data type - arrays
```

```python
# generate data

x = np.linspace(-5, 5, 100)

x
```

```python
f = np.sin(x)

f
```

The first approach - functional/matlab style

```python
plt.figure()
plt.plot(x,f)
plt.title("sin(x)")
plt.xlabel("x")
plt.ylabel("f")
```

The second approach - OOP

```python
fig = plt.figure() # create one figure
ax = plt.axes() #on the figure, create an ax
ax.plot(x,f) #on the axe, create a plot
ax.set(xlim=(-10,10),ylim=(-5,5),xlabel="x", ylabel="y",title="sin(x)")
```
